```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
def count_pos_tags(text):
    """
    Count the number of verbs, nouns, pronouns, and adjectives in the given text.

    Args:
    - text: The input text to analyze.

    Returns:
    - A dictionary containing the counts of each part of speech.
    """
    # Tokenize the text into words
    words = word_tokenize(text)

    # Tag the words with their corresponding part of speech
    tagged_words = pos_tag(words)

    # Initialize counters
    counts = {
        'verb': 0,
        'noun': 0,
        'pronoun': 0,
        'adjective': 0
    }

    # Count the occurrences of each part of speech
    for word, tag in tagged_words:
        if tag.startswith('VB'):
            counts['verb'] += 1
        elif tag.startswith('NN'):
            counts['noun'] += 1
        elif tag.startswith('PRP'):
            counts['pronoun'] += 1
        elif tag.startswith('JJ'):
            counts['adjective'] += 1

    return counts
```

```python
# Test case 1: Analyzing a simple phrase
phrase = "The quick brown fox jumps over the lazy dog."
result = count_pos_tags(phrase)
print("Counts for test case 1:")
print(result)
```

```python
# Test case 2: Analyzing a paragraph
paragraph = "NLP, or Natural Language Processing, is a field of study focused on enabling computers to understand and interpret human language. It involves tasks such as text classification, sentiment analysis, and named entity recognition."
result = count_pos_tags(paragraph)
print("\nCounts for test case 2:")
print(result)
```
