```python
data['Text'] = data['Text'].apply(lambda x: x.strip())
data['Label'] = '__label__' + data['Language']
data[['Label', 'Text']].to_csv('preprocessed_data.txt', sep='\t', index=False, header=False)
```

```python
data.head()
```

```python
import fasttext

# Step 3: Train the FastText model
train_data = 'preprocessed_data.txt'
learning_rate = 0.1
epoch_count = 10
dimension = 100

# Train the model
model = fasttext.train_supervised(input=train_data, lr=learning_rate, epoch=epoch_count, dim=dimension)
```

```python
# Step 5: Evaluate the model
eval_data = 'preprocessed_data.txt'

# Evaluate the model
evaluation = model.test(eval_data)
accuracy = evaluation[1]

print('Accuracy:', accuracy)
```

```python
# Step 6: Use the model for language detection
text_sample = "Hello, how are you?"  # Text sample in an unknown language

# Predict the language of the text sample
prediction = model.predict(text_sample)

# Extract the predicted language
predicted_language = prediction[0][0].replace('__label__', '')

print('Predicted language:', predicted_language)
```
