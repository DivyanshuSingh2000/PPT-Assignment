```python
import requests
from bs4 import BeautifulSoup

# Specify the URL of the Amazon webpage you want to scrape
url = "https://www.amazon.com/"

# Send a GET request to the webpage
response = requests.get(url)

# Parse the HTML content of the webpage
soup = BeautifulSoup(response.content, 'html.parser')

# Find all image tags on the webpage
image_tags = soup.find_all('img')

# Iterate over the image tags and download the images
for i, img in enumerate(image_tags):
    # Check if the 'src' attribute exists
    if 'src' in img.attrs:
        image_url = img['src']
        # Send a GET request to download the image
        image_response = requests.get(image_url)
        # Save the image to a local directory
        with open(f'image_{i}.jpg', 'wb') as f:
            f.write(image_response.content)
    else:
        print(f"Image {i} doesn't have a 'src' attribute.")
```
